```python
def label_dir(from_directory, to_directory, label_map):
    i=0
    if not os.path.isdir(to_directory):
        os.mkdir(to_directory)
    for filename in os.listdir(from_directory):
        try:
            label = label_map[filename[:-4]]
            path_dir = os.path.join(to_directory, label)
            if not os.path.isdir(path_dir):
                os.mkdir(path_dir)
            file_path_from = os.path.join(from_directory, filename)
            file_path_to = os.path.join(path_dir, filename)
            if not os.path.exists(file_path_to):
                shutil.copyfile(file_path_from, file_path_to)
        except:
            print("Exception for index: ", i, "Breed: ", label, ", File: ", file_path_from)
        i+=1
        if i%1000==0:
            print(i)
```
